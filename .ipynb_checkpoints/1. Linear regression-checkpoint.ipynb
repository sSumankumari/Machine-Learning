# Linear Regression with single variable

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_excel("areaplot.xlsx")
df
```

```python
%matplotlib inline
plt.xlabel('Area(sqr ft)')
plt.ylabel('Price(US$)')
plt.scatter(df.area, df.price, color='orange', marker='*')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']], df.price) # training the model
```

```python
reg.predict([[3300]])
```

```python
# Value for slope or gradient 'm'
reg.coef_
```

```python
# Value for intercept 'b' in (y=mx+b)
reg.intercept_
```

```python
# y = m*x + b
135.78767123 * 3300 + 180616.43835616432
```

```python
%matplotlib inline
plt.xlabel('Area(sqr ft)', fontsize=12)
plt.ylabel('Price(US$)', fontsize=12)
plt.scatter(df.area, df.price, color='orange', marker='*')

plt.plot(df.area, reg.predict(df[['area']]), color='blue')
```

```python
data = pd.read_excel("testingarea.xlsx")
data.head(3)
```

```python
predictedPrice = reg.predict(data)
predictedPrice
```

```python
data['prices'] = predictedPrice
```

```python
data
```

```python
# data.to_csv("prediction.csv", index=False)
data.to_excel("prediction.xlsx")
```

## Build a regression model and predict the per capita income fo canadian citizens in year 2020

```python
df = pd.read_csv("canada_per_capita_income.csv")
df
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(df.year, df[['per capita income (US$)']], color='green', marker='*')
```

```python
new = df.drop('per capita income (US$)', axis='columns')
new
```

```python
income = df[['per capita income (US$)']]
income
```

```python
# creating linear regression object
reg = linear_model.LinearRegression()
reg.fit(new, income)
```

### Predicting the per capita income in year 2020

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[2020]])
```
