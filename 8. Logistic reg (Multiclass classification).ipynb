# Logistic Regression (Multiclass Classification)

```python
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
digits = load_digits()
```

```python
dir(digits)
```

```python
digits.data[0]
```

```python
plt.gray()
for i in range(5):
    plt.matshow(digits.images[i])
```

```python
digits.target[0:5]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.2)
```

```python
len(x_train)
len(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
plt.matshow(digits.images[67])
```

```python
digits.target[67]
```

```python
model.predict([digits.data[67]])
```

```python
model.predict(digits.data[0:5])
```

### Creating Confusion Matrix

```python
y_predicted = model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predicted)
cm
```

```python
# Confusion Matrix visualization
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

#### Using sklearn.datasets iris flower dataset to train the model using logistic regression. And also to figure out the accuracy of model and use that to predict different samples in your test dataset. In iris dataset there are 150 samples containing following features,

1. Sepal Length
2. Sepal Width
3. Petal Length
4. Petal Width

#### Using above 4 features also clasify a flower in one of the three categories,

1. Setosa
2. Versicolour
3. Virginica

```python
import pandas as pd
iris_data = pd.read_csv('IRIS.csv')
iris_data
```

```python
final_data = iris_data.drop(['species'], axis='columns')
final_data
```

```python
x = final_data
y = iris_data.species
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
flower_model = LogisticRegression(max_iter=1000)
flower_model.fit(x_train, y_train)
```

```python
x_train
```

```python
accuracy = flower_model.score(x_test, y_test)
accuracy
```

```python
flower_model.predict(x_test)
```
